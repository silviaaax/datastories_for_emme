```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
accidents = pd.read_csv("/content/drive/MyDrive/injuries.zip", encoding="latin-1")
```

```python
accidents.head()
```

```python
accidents.drop(["ID","UPA","Address1","Address2","Primary NAICS","Inspection",
         "Nature","Part of Body","Event","Source","Secondary Source"],axis=1,inplace=True)
```

```python
accidents['State'].value_counts().mean()
```

Οι 10 πρώτες πολιτείες με τα περισσότερα ατυχήματα

```python
accidents['State'].value_counts().head(10).plot(kind='bar' , color=['darkorange', 'b'])
```

Οι 10 πρώτες πόλεις με τα περισσότερα ατυχήματα

```python
accidents['City'].value_counts().head(10).plot(kind='bar' , color=['purple', 'cyan'])
```

Τα 10 πιο συχνά ατυχήματα

```python
accidents['EventTitle'].value_counts().head(10).plot(kind='barh' , color=['powderblue', 'orchid'])
```

Οι  10 πιο συχνές αιτίες των ατυχημάτων

```python
accidents['SourceTitle'].value_counts().head(10).plot(kind='barh' , color=['teal', 'darkorchid'])
```

Τα 10 πιο συχνα μέρη του σώματος που τραυματίζονται στα ατυχήματα

```python
accidents['Part of Body Title'].value_counts().head(10).plot(kind='barh' , color='lightsalmon')
```

Οι πολιτειες με τους περισσότερους νοσηλευόμενους μετά απο ατύχημα

```python
accidents.groupby('State')['Hospitalized'].sum().sort_values(ascending = False).head().plot(kind='bar' , color='crimson')
```

Οι πόλεις με τα περισσότερα ατυχήματα από την πολιτεία με τα περισσότερα ατυχήματα

```python
accidents[accidents['State'] == 'TEXAS']['City'].value_counts().head().plot(kind='bar', color='lime')
```

Οι πιο συχνές ημερομηνίες ακρωτηριασμών

```python
accidents.groupby('EventDate')['Amputation'].sum().sort_values(ascending=False).head().plot(kind='bar' , color=['darkorange', 'darkorchid'])
```

Οι  10 εργοδότες με το μεγαλύτερο άθροισμα ατυχημάτων

```python
accidents.groupby('Employer')['Hospitalized'].sum().sort_values(ascending = False).head(10).plot(kind='bar' , color=['coral', 'lime'])
```

Οι αιτίες ατυχημάτων στην εταιρεία με τα περισσότερα ατυχήματα

```python
accidents[accidents['Employer'] == 'U.S Postal Service']['SourceTitle'].value_counts().head().plot(kind='bar', color='plum')
```
